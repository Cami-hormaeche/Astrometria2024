```python
df_train.set_index('ID',inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_tr, df_te=train_test_split(df_train,test_size=0.3,random_state=42)
#Divide el conjunto de datos en entrenamiento (70%) y prueba (30%
```

```python
X_train=df_tr.drop('Hclass',axis=1)
Y_train=df_tr['Hclass']
# Separa las variables independientes (X_train, X_test) de la dependiente (Y_train, y_test).
X_test=df_te.drop('Hclass',axis=1)
y_test=df_te['Hclass']
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(X_train,Y_train)
#Entrena un modelo de árbol de decisión con los datos de entrenamiento.
```

```python
y_pred=clf.predict(X_test)
#Realiza predicciones en los datos de prueba.
```

```python
from sklearn.metrics import f1_score,confusion_matrix,classification_report
```

```python
confusion_matrix(y_test,y_pred) #Clase y prediccion, si la prediccion es buena deberia dar una diagonal
#Muestra la matriz de confusión para analizar errores.
```

```python
print(classification_report(y_test,y_pred)) 
#MUestra para cada calse cual es la precision del ricall y el f1-score
#Hace promedio pesado
#La metrica que evalua la competencia es la media, promedio pesado
#Proporciona métricas detalladas como precisión, recall y F1-score para cada clase.
```

```python
df_submit=pd.read_csv('test.csv')
df_submit.set_index('ID',inplace=True)
```

```python
prediction= clf.predict(df_submit)
```

```python
df_submit['Hclass']=prediction
```

```python
df_submit.loc[:,['Hclass']].to_csv('submission.csv',index=True) #Este es el que queremos subir a la competencia
#El besline es este
#En el publico subimos el porcentaje de conjunto de test
#Carga el conjunto de prueba final.
#Predice las clases utilizando el modelo entrenado.
#Guarda las predicciones en un archivo submission.csv para la competencia.
```

```python
f1_score(y_test,y_pred,average='weighted')
```

```python
param_grid={
    'max_depth':[5,10,15,20,25,30,35,40,45,50],
    #'min_samples_split':[2,5,10,15,20,25,30,35,40,45,50] #podriamos agregar esto ademas
    } #Diccionario, entre '' claves y el valor que se le asigna a la llave o clave entre []
#En este caso particular definimos los parametros del modelo de esta manera entonces entre [] tenemos que agregar la lista correspondiente al hiper parametro, en este caso un entero
```
