La idea de este ejercicio es graficar la magnitud en la banda r como función de la magnitud en la banda g y luego realizar un ajuste lineal a los datos para encontrar la relación entre ambas magnitudes.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datos = pd.read_csv('Galaxias_color_2.csv')
```

```python
# Filtrar los datos para evitar magnitudes negativas
datos_filtrados = datos[(datos['petroMag_g'] > 0) & (datos['petroMag_r'] > 0)]
```

```python
#Filtramos por tipo de galaxias
datos_elipticas = datos_filtrados[datos_filtrados['elliptical'] == 1]

datos_espirales = datos_filtrados[datos_filtrados['spiral'] == 1]
```

```python
#creamos el grafico de galaxias elipticas primero 
sns.scatterplot(data=datos_elipticas,x='petroMag_g', y='petroMag_r', color='darkturquoise')
plt.title('Galaxias Elípticas: Magnitudes Aparentes de r vs g')
plt.xlabel('Magnitud en g')
plt.ylabel('Magnitud en r')
```

```python
#Ahora, observando la tendencia de los datos queremos realizar un ajuste lineal, dado que los datos tienen esta tendencia
slope, intercept = np.polyfit(datos_elipticas['petroMag_g'], datos_elipticas['petroMag_r'], 1)
regression_line = slope * (datos_elipticas['petroMag_g']) + intercept
print(regression_line)
```

```python
#Ploteamos para galaxias elipticas los datos de magnitud g vs r y realizamos el ajuste lineal
#copiamos lo del primer grafico
sns.scatterplot(data=datos_elipticas,x='petroMag_g', y='petroMag_r',color='darkturquoise')

#Ahora ploteamos el ajuste lineal
plt.plot(datos_elipticas['petroMag_g'], regression_line, color='red', label=f'Ajuste lineal: Mag_r = {slope:.2f} * Mag_g + {intercept:.2f}')
plt.title('Galaxias Elípticas: Magnitudes Aparentes de r vs g con Ajuste Lineal')
plt.xlabel('Magnitud en g')
plt.ylabel('Magnitud en r')
plt.legend()
plt.savefig('Galaxias Elipticas: Magnitudes Aparentes de r vs g con Ajuste Lineal', dpi=300, bbox_inches='tight')
plt.show()
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
```

Repetimos para galaxias espirales lo mismo que para elipticas


```python
slope, intercept = np.polyfit(datos_espirales['petroMag_g'], datos_espirales['petroMag_r'], 1)
regression_lineal_2 = slope * (datos_espirales['petroMag_g']) + intercept
print(regression_lineal_2)
```

```python
#ploteamos y realizamos el ajuste para las espirales
sns.scatterplot(data=datos_espirales,x='petroMag_g', y='petroMag_r')

#Ahora ploteamos el ajuste lineal
plt.plot(datos_espirales['petroMag_g'], regression_lineal_2, color='red', label=f'Ajuste lineal: Mag_r = {slope:.2f} * Mag_g + {intercept:.2f}')
plt.title('Galaxias Espirales: Magnitudes Aparentes de r vs g con Ajuste Lineal')
plt.xlabel('Magnitud en g')
plt.ylabel('Magnitud en r')
plt.legend()
plt.savefig('Galaxias Espirales: Magnitudes Aparentes de r vs g con Ajuste Lineal', dpi=300, bbox_inches='tight')
plt.show()
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
```
