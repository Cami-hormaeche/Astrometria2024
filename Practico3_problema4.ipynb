La idea de este ejercicio es graficar la magnitud en la banda r como función de la magnitud en la banda g y luego realizar un ajuste lineal a los datos para encontrar la relación entre ambas magnitudes.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datos = pd.read_csv('Galaxias_color_2.csv')
```

```python
# Filtrar los datos para evitar magnitudes negativas
datos_filtrados = datos[(datos['petroMag_g'] > 0) & (datos['petroMag_r'] > 0)]
```

```python
# Filtrar galaxias elípticas
datos_elipticas = datos_filtrados[datos_filtrados['elliptical'] == 1]

# Filtrar galaxias espirales
datos_espirales = datos_filtrados[datos_filtrados['spiral'] == 1]
```

```python
# Crear gráficos para galaxias elípticas
plt.figure(figsize=(10, 6))
sns.scatterplot(data=datos_elipticas, x='petroMag_g', y='petroMag_r', alpha=0.6)
plt.title('Galaxias Elípticas: Magnitudes Aparentes de r vs g')
plt.xlabel('Magnitud en g')
plt.ylabel('Magnitud en r')

# Ajustar la regresión lineal para elípticas
x_elipticas = datos_elipticas['petroMag_g']
y_elipticas = datos_elipticas['petroMag_r']
coef_elipticas = np.polyfit(x_elipticas, y_elipticas, 1)
polynomial_elipticas = np.poly1d(coef_elipticas)

# Crear valores para el ajuste
x_fit_elipticas = np.linspace(min(x_elipticas), max(x_elipticas), 100)
y_fit_elipticas = polynomial_elipticas(x_fit_elipticas)

# Graficar la línea de ajuste para elípticas
plt.plot(x_fit_elipticas, y_fit_elipticas, color='red', label='Ajuste Lineal: r = {:.2f}g + {:.2f}'.format(coef_elipticas[0], coef_elipticas[1]))
plt.legend()
plt.grid()
plt.show()

# Imprimir la ecuación de la línea de ajuste para elípticas
print(f"Ecuación de ajuste (Elípticas): r = {coef_elipticas[0]:.2f} * g + {coef_elipticas[1]:.2f}")
```

```python

# Crear gráficos para galaxias espirales
plt.figure(figsize=(10, 6))
sns.scatterplot(data=datos_espirales, x='petroMag_g', y='petroMag_r', alpha=0.6)
plt.title('Galaxias Espirales: Magnitudes Aparentes de r vs g')
plt.xlabel('Magnitud en g')
plt.ylabel('Magnitud en r')

# Ajustar la regresión lineal para espirales
x_espirales = datos_espirales['petroMag_g']
y_espirales = datos_espirales['petroMag_r']
coef_espirales = np.polyfit(x_espirales, y_espirales, 1)
polynomial_espirales = np.poly1d(coef_espirales)

# Crear valores para el ajuste
x_fit_espirales = np.linspace(min(x_espirales), max(x_espirales), 100)
y_fit_espirales = polynomial_espirales(x_fit_espirales)

# Graficar la línea de ajuste para espirales
plt.plot(x_fit_espirales, y_fit_espirales, color='red', label='Ajuste Lineal: r = {:.2f}g + {:.2f}'.format(coef_espirales[0], coef_espirales[1]))
plt.legend()
plt.grid()
plt.show()

# Imprimir la ecuación de la línea de ajuste para espirales
print(f"Ecuación de ajuste (Espirales): r = {coef_espirales[0]:.2f} * g + {coef_espirales[1]:.2f}")


```
